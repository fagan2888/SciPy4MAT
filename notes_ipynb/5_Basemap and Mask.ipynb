  + lon and lat labels

### Masking arrays

The plot drawn above has some bad values. The MATLAB approach to removing these values from the plot would be to "NaN" the bad values. The Python approach is to mask the values. This is done using a `numpy.masked_array`. Marplotlib makes use of masked arrays and does not plot masked values. This is useful as we do not have to change any of the values that we're working with.
